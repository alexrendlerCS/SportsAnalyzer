```python
from nba_api.stats.endpoints import playergamelog
from nba_api.stats.library.parameters import SeasonAll
from nba_api.stats.endpoints import commonallplayers

# Specify the player's full name
player_full_name = 'Jalen Duren'

# Fetch the list of all players
all_players = commonallplayers.CommonAllPlayers(is_only_current_season=1)

# Convert the data to a dictionary
all_players_data = all_players.get_normalized_dict()

# Find the player ID by searching for the player's full name
for player_info in all_players_data['CommonAllPlayers']:
    if player_info['DISPLAY_FIRST_LAST'] == player_full_name:
        player_id = player_info['PERSON_ID']
        print(f"The player ID for {player_full_name} is {player_id}")
        break
else:
    print(f"Player {player_full_name} not found.")
season = '2023-24'

# Fetch the game log for the specified player and season
gamelog = playergamelog.PlayerGameLog(player_id=player_id, season=season, season_type_all_star='Regular Season')

# Convert the data to a dictionary
gamelog_data = gamelog.get_normalized_dict()

# Get the game log entries
game_entries = gamelog_data['PlayerGameLog']

# Print the statistics for each game
for game in game_entries:
    print(f"Game Date: {game['GAME_DATE']}, Points: {game['PTS']}, Rebounds: {game['REB']}, Assists: {game['AST']}")
```
